```python
import cv2
import numpy as np 

image2=cv2.imread('7.jpg')
hsv = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
#cv2.imshow('HSV',hsv)

contours,hierarchy = cv2.findContours(hsv,cv2.RETR_TREE ,cv2.CHAIN_APPROX_NONE)
#cv2.drawContours(image2, contours, -1, (0,255,0), 3)
#cv2.imshow('Contours',image2)
print ("Number of contours = ", len(contours))

areas = [cv2.contourArea(c) for c in contours]

max_index = np.argmax(areas)
print('index: ', max_index)
cnt=contours[max_index]

epsilon = 0.05*cv2.arcLength(cnt,True)
approx = cv2.approxPolyDP(cnt,epsilon,True)
sides=approx.size

print('SIZE: ', sides)
x,y,w,h = cv2.boundingRect(cnt)
#rect=cv2.rectangle(image2,(x,y),(x+w,y+h),(0,255,0),2)
crop_img = image2[y:y+h, x:x+w]
output = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
print(output)
cv2.imshow('Rect', crop_img)
print ('Number of pixel in image: ',output.size)

pixel=cv2.countNonZero(output)
print('countNonZero:',pixel)
print('Ratio: ', (pixel/output.size))
cv2.waitKey()
cv2.destroyAllWindows()
```
